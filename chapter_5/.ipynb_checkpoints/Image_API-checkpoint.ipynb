```python
#resize : tf.image.resize_area,tf.image.resize_bicubic,tf.image.resize_nearest_neighbor
# ,

name = './fengjing.jpg'
img_str = tf.read_file(name)  # 
img_decode = tf.image.decode_image(img_str) # 
img_decode = tf.reshape(img_decode,[1,1200,1920,3])

resize_img = tf.image.resize_bicubic(img_decode,[2400,3840])
resize_img_reshape = tf.reshape(resize_img,[2400,3840,3])
sess = tf.Session()
img_decode_val = sess.run(resize_img_reshape)  #img_decode_valfloatimshow()int
img_decode_val = np.asarray(img_decode_val,np.uint8)



print(img_decode_val.shape) 
print (image_val.shape)
%matplotlib inline
imshow(img_decode_val)

```

```python
#crop 
name = './fengjing.jpg'
img_str = tf.read_file(name)  # 
img_decode = tf.image.decode_image(img_str) # 
img_decode = tf.reshape(img_decode,[1,1200,1920,3])

padded_img = tf.image.pad_to_bounding_box(img_decode,
                                         50,100, #
                                         2000,3000) #
padded_img_reshape = tf.reshape(padded_img,[2000,3000,3])
sess = tf.Session()
img_decode_val = sess.run(padded_img_reshape) 
img_decode_val = np.asarray(img_decode_val,np.uint8)



print(img_decode_val.shape) 
print (image_val.shape)
%matplotlib inline
imshow(img_decode_val)
```

```python
#flip : 
#tf.image.flip_up_down,
#tf.image.flip_left_right,
#tf.image.random_filp_up_down,#
#tf.image.random_filp_left_right

name = './fengjing.jpg'
img_str = tf.read_file(name)  # 
img_decode = tf.image.decode_image(img_str) # 
img_decode = tf.reshape(img_decode,[1,1200,1920,3])
fliped_img = tf.image.flip_up_down(img_decode)
                                         
fliped_img_reshape = tf.reshape(fliped_img,[1200,1920,3])
sess = tf.Session()
img_decode_val = sess.run(fliped_img_reshape) 
img_decode_val = np.asarray(img_decode_val,np.uint8)
print(img_decode_val.shape) 
print (image_val.shape)
%matplotlib inline
imshow(img_decode_val)
```
