```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = np.random.rand(100)
noise = np.random.normal(0,0.01,x_data.shape)
y_data = 0.1*x_data+0.2+noise
plt.scatter(x_data,y_data)
plt.show()
```

```python
line_model = Sequential()
line_model.add(Dense(1,input_shape=(1,)))
line_model.compile(optimizer=SGD(),loss="MSE")
```

```python
for step in range(5001):
    cost = line_model.train_on_batch(x_data,y_data)
    if step%500 == 0:
        print("step:",step,"cost:",cost)
        
```

```python
w,b = line_model.layers[0].get_weights()
print("w:",w,"b:",b)
```

```python
y_pre = line_model.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pre,"r")
plt.show()
```

```python
x_data = np.linspace(-1,1,200)
noise = np.random.normal(0,0.1,x_data.shape)
y_data = np.square(x_data)+noise
plt.scatter(x_data,y_data)
plt.show()
```

```python
line_model = Sequential()
line_model.add(Dense(5,input_shape=(1,),activation="tanh"))
line_model.add(Dense(3,))
line_model.add(Dense(1,))

line_model.compile(optimizer=SGD(),loss="MSE")
for step in range(5001):
    cost = line_model.train_on_batch(x_data,y_data)
    if step%500 == 0:
        print("step:",step,"cost:",cost)
w,b = line_model.layers[0].get_weights()
print("w:",w,"b:",b)
y_pre = line_model.predict(x_data)
plt.scatter(x_data,y_data)
plt.plot(x_data,y_pre,"r-",lw=3)
plt.show()
```
