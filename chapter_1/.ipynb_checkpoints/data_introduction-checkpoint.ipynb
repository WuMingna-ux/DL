```python
import pickle as pk
import numpy as np
import os

CIFAR_DIR ='../../datasets/cifar-10-batches-py'
dirs = os.listdir( CIFAR_DIR )
print(dirs)
```

```python
with open(os.path.join(CIFAR_DIR, 'data_batch_1'),'rb') as f:
    
    data = pk.load(f,encoding='iso-8859-1')
    print(type(data))
    print(data.keys())
    print((data['batch_label'])) # 文件的含义 ，5个data_batch中的一个
    print(type(data['filenames'])) #图片对应的名字 ^eg: ***.png
    print(type(data['labels'])) #标签
    print(type(data['data'])) #存储像素
    print(data['data'].shape)
    # 32 * 32 * 3 = 3072 
    # RR ->62 * 32 --->RR-GG-BB -->32 * 32 * 3 = 3072 
```

```python
image_arr = data['data'][100]
image_arr = image_arr.reshape((3,32,32)) # 第一个参数是3个通道，32*32是每个通道的像素值
image_arr = image_arr.transpose((1,2,0)) #通道的交换
print(image_arr.shape)

import matplotlib.pyplot as plt
image = plt.imshow(image_arr)
plt.show(image)
```
